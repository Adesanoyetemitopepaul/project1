```python
def wrangle(filepath):
    
    #Read CSV file and set index as timestamp
    df=pd.read_csv(filepath)
    
    # convert the timestamp column to datetime
    df["timestamp"] = pd.to_datetime(df['timestamp'])
    
    # use that column as an index for the whole df;
    df.set_index('timestamp',drop=False, inplace=True)
    
    df.index = pd.DatetimeIndex(df.index).to_period('H') #this is use to convert time stamp to hrs
    
    # when i tried localizing the timezone to UTC, it said timezone aware, so i just converted to Africa?Nairobi straight
    #df.index  = df.index.tz_localize("UTC").tz_convert('Africa/Nairobi')
    
    df.drop(columns=["sensor_id","sensor_type","location","lat","lon",],inplace=True)
    
    #we need p2 only
    mask = df["value_type"]=="P2"
    df=df[mask]
    
    #remove outliers
    df=df[df["value"]<200]
    
    df.drop(columns=["value_type"],inplace=True)
    
    
    #Resampling the time series, filling for missing values and converting  to series 
    
    df=df["value"].resample("1H").mean().fillna(method="ffill")#.to_frame()
    
    #i am dropping value type because its a string and it wont work for calculation and i aslo dont need it for prediction
    return df
```

```python
#ARMA MODEL IS LOOKING AT PAST READING TO MAKE PRESENT PREDICTION WHICH IS P WHILE THE q IS THE PAST PREDICTION, SPECIFICALLY ERROR in past prediction. so ARMA MODEL cover for difference between our true value and past prediction at previous timestamp(ERROR).this cover for not only for what was the reading at prevoius timestamp but how far off was my reading at previous estimate

#AR(autoregresion p)
#MA(moving average)



#AR MODEL IS USING PAST READING TO MAKE PRESENT PREDICTION
#P= LAG( NUMBER OF STEp WE NEED TO GO TO THE PAST)
#ARMA MODEL= WE DONT JUST LOOK AT PAST READING BUT ALSO PAST PREDICTION
#MA MOVING AVERAGE
#IN ARMA MODEL WE HAVE TO DECIDE WHAT P AND Q IS 
#P IS LAG, Q IS ERROR LAG. P AND Q ARE HYPERPARAMETER.
#WHY HYPERPARAMETER,ITS BECCAUSE BEFORE WE FIT OUR MODEL WITH DATA WE MAKE DECISION  ON HOW MANY P AND Q TO HAVE IN OUR EQUATION. 
#THIS WILL DETERMINE OUR MODEL IS TRAINED

#use your pacf to determine p 
#use your acf to determine q
#or we can do hyperparameter tunning to determine p and q which is a better way
#p_params =range(0,25,8)
#q_params =range(0,3,1)

#why are we using MA (moving average) term, because time series model suffer what is called  an exergerated shock which is a one time event  that throw the reading for subsequent timestamp out of normal. 
#if there are fire work or collpse of building etc  e.g on independent day on monday subsequently there will be more air pollution on tuesday, wednesday etc, this will have shock or effect on subsequnt reading this is called exergerated shock so we try to us MA(moving avereage) to correct that
```

```python
#Task 3.4.8: Create ranges for  possible p and  ùëû values. p_params should range between 0 and 25, by steps of 8. q_params should range between 0 and 3 by steps of 1.
#use your pacf to determine p 
#use your acf to determine q
#or we can do hyperparameter tunning.
p_params =range(0,25,8)
q_params =range(0,3,1)

#less q than p is because q is quick short term shock into the system so we want them small.
```

```python
#Task 3.4.9: Complete the code below to train a model with every combination of hyperparameters in p_params and q_params. Every time the model is trained, the mean absolute error is calculated and then saved to a dictionary. 
#If you're not sure where to start, do the code-along with Nicholas!

#after ploting( or train a model) by putting p and q together wit p on horizontal roll as (0,8,16,24) and q(0,1,2) on the vertical row, 
#like probability head and toe plotting. the  model will be (0,0),(0,1)(0,2)(8,0),(8,1),(8,2) etcafter training the model with those hyperparameter we will need to, calculate training Mae for each of the model and save the result,  we will need to do hat for every  square in each grid, once we are done will look for every mae and find the one with best perfomance
#that is what the code below explain


# Create empty dictionary to store MAEs
mae_grid = dict()
# Outer loop: Iterate through possible values for `p`
for p in p_params:
    # Create key-value pair in dict. Key is `p`, value is empty list.
    mae_grid[p] = list()
    # Inner loop: Iterate through possible values for `q`
    for q in q_params:
        # Combination of hyperparameters for model
        order = (p, 0, q) # arima have three order so we are keeping the middle one blanck.
        # Note start time
        start_time = time.time() #we can time it to give us time to train so that it will not be too slow, we time it, to know the time it take to train model
        # Train model
        model =ARIMA(y_train, order=order).fit()
        # Calculate model training time
        elapsed_time = round(time.time() - start_time, 2) 
        print(f"Trained ARIMA {order} in {elapsed_time} seconds.")
        # Generate in-sample (training) predictions
        df_pred = model.predict()
        # Calculate training MAE
        mae = mean_absolute_error(df_train, df_pred)
        # Append MAE to list in dictionary
        mae_grid[p].append(mae)

print()
print(mae_grid)
```
